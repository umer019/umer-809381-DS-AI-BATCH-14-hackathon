```python

choice=''
#WHILE LOOP LAGAYA TAAKE YE PROGRAM MULTIPLE TIMES REPEAT HO JAB TK USER NA CHAHE (4 = EXIT )
while choice != "4":
    
    print("\n===== QUIZ MENU =====")
    print("1. Start Quiz")
    print("2. View Last_Score")
    print("3. Performance Analysis")
    print("4. Exit")
    choice=(input('enter your choice'))
# IF / ELSE : AGR USER 1 CHOOSE KRTA HAI TO YE 10 QUESTIONS POOCHNE HAI     
    if choice == "1":
    # ABHI USKA SCORE  0 HAI TO SCORE KO ISLIYE O SE ASSIGN KRADYA    
        score = 0
        
        questions = [
            ["Which keyword is used to define a function in Python?", "A) func", "B) function", "C) def", "D) lambda", "C"],
            
            ["What is the output of print(type([]))?", "A) <class 'list'>", "B) <class 'tuple'>", "C) <class 'dict'>", "D) <class 'set'>", "A"],
            
            ["Which of the following is immutable in Python?", "A) Set", "B) List", "C) Tuple", "D) Dictionary", "C"],
            
            ["What does the len() function return?", "A) The type of an object", "B) The length of an object", "C) The last element", "D) The maximum value", "B"],
            
            ["Which operator is used for floor division in Python?", "A) /", "B) **", "C) %", "D) //", "D"],
            
            ["What is the correct way to import the math module?", "A) require math", "B) using math", "C) include math", "D) import math", "D"],
            
            ["Which of the following is used to handle exceptions in Python?", "A) if-else", "B) for-while", "C) try-except", "D) switch-case", "C"],
            
            ["What is the output of bool(0)?", "A) False", "B) 0", "C) None", "D) True", "A"],
            
            ["Which method is used to add an element to a list?", "A) add()", "B) insert()", "C) extend()", "D) append()", "D"],
            
            ["What is the output of print(2 ** 3)?", "A) 6", "B) 8", "C) 16", "D) 9", "B"]
        ]
        
        i = 0

        # question loop ; har question aik aik baar print hoga jiske sath options bhi honge A,B,C,D 
        while i < len(questions):
            
            print("\n" + questions[i][0])
            print(questions[i][1])
            print(questions[i][2])
            print(questions[i][3])
            print(questions[i][4])
            # ANS ka input A,B,C,D MAIN
            ans = input("Enter your answer (A/B/C/D): ")
            #AGR USER NE GALAT INPUT DIYA MATLB A,B,C,D KE ILAWA KUCH OR TO WO WHILE LOOP KE ZARIYE DUBARA REPEAT HOGA OR USER KO DUBARA INPUT DALNE KA KKAHE GA 
            while ans not in ["A", "B", "C", "D"]:
                ans = input("Invalid! Enter A/B/C/D: ")
            # AGR ANS SAHI HAI TO TOTAL SCORE MAIN 5 ADD HOGA OR AGR GALAT ANSWER DIYA TO ANS MAIN 5 MINUS HOGA
            if ans == questions[i][5]:
                print("Correct ")
                score = score + 5
            else:
                print("Wrong ")
                score = score - 5
            
            i = i + 1
        #LAST SCORE MAIN ADD ISLIYE KYA , AGR USER NE DOBARA POOCHA KE LAST SCORE KYA THA USKA 
        last_score = score
        print("\nYour Score:", score)
    
    elif choice == "2":
        print("\nLast Score:", last_score)
    
    elif choice == "3":
        
        print("\n--- Performance ---")
        
        if last_score <= 0:
            print("Level: Beginner")
            print("Feedback: Work harder!")
        elif last_score <= 2:
            print("Level: Intermediate")
            print("Feedback: Not bad ")
        else:
            print("Level: Advanced")
            print("Feedback: Excellent ")
    
    elif choice == "4":
        print("Goodbye ")
    
    else:
        print("Invalid choice! Try again.")
```
